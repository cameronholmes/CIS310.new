```python
d = {'col1': [1,2,4,3,2,1,6,7,3,9], 'col2': [5,8,9,10,12,8,6,2,8,7]}
df2 = pd.DataFrame(d)
df2
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(df2['col1'], 'k.', label='col1')
ax.plot(df2['col2'], 'ko--', label='col2')
ax.legend(loc='best')
```
